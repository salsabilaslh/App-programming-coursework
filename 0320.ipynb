```python
from faker import Faker
import pandas as pd
```

```python
class Person:
    def __init__(self, name, age, address, phone, job):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone
        self.job = job
```

```python
fake = Faker()
```

```python
people = []

for i in range(1000):
    person = Person(
        fake.name(),
        fake.random_int(min=18, max=60),
        fake.address(),
        fake.phone_number(),
        fake.job()
    )

    people.append({
        "Name": person.name,
        "Age": person.age,
        "Address": person.address,
        "Phone": person.phone,
        "Job": person.job
    })
```

```python
df = pd.DataFrame(people)
```
