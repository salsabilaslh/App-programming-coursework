# 📚 Web Scraping Project: Books to Scrape Analysis

## 📖 Introduction
This project aims to perform web scraping using BeautifulSoup on the "Books to Scrape" website.
The goal is to extract book titles, prices, and categories, and analyze the most expensive books.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://books.toscrape.com/"
res = requests.get(url)

print(res.status_code)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

## 🔹 Task 1: Extract First 20 Book Titles

This section extracts the first 20 book titles from the main page.

```python
books = soup.find_all('h3')

titles = []
for book in books:
    title = book.find('a')['title']
    titles.append(title)

df_20 = pd.DataFrame(titles, columns=['Title'])
df_20
```

## 🔹 Task 2: Extract 100 Book Titles from Multiple Pages

This section collects book titles from the first 5 pages (100 books total).

```python
titles_100 = []

for page in range(1, 6):
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    books = soup.find_all('h3')

    for book in books:
        title = book.find('a')['title']
        titles_100.append(title)

df_100 = pd.DataFrame(titles_100, columns=['Title'])
df_100
```

```python
categories = {
    "Travel": "travel_2",
    "Mystery": "mystery_3",
    "Historical Fiction": "historical-fiction_4",
    "Sequential Art": "sequential-art_5",
    "Classics": "classics_6",
    "Philosophy": "philosophy_7",
    "Romance": "romance_8",
    "Womens Fiction": "womens-fiction_9",
    "Fiction": "fiction_10",
    "Childrens": "childrens_11"
}
```

## 🔹 Task 3: Extract Book Data by Category (Title, Price, Category)

This section scrapes book data from 10 categories, including title, price, and category information.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

data = []

for cat_name, cat_url in categories.items():
    url = f"http://books.toscrape.com/catalogue/category/books/{cat_url}/index.html"

    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    books = soup.find_all('article', class_='product_pod')

    for book in books:
        title = book.h3.a['title']

        price_text = book.find('p', class_='price_color').text
        price = float(re.sub(r'[^0-9.]', '', price_text))

        data.append([cat_name, title, price])

df = pd.DataFrame(data, columns=['Category', 'Title', 'Price'])
df
```

## 💰 Task 4: Top 10 Most Expensive Books

The data is sorted by price in descending order to identify the top 10 most expensive books.

```python
top10 = df.sort_values(by='Price', ascending=False).head(10)
top10
```
