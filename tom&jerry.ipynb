```python
class User:
    def __init__(self, name):
        self.name = name
        self.following = 0
        self.followers = 0

    def follow(self, other):
        self.following += 1
        other.followers += 1

        print(f"{self.name} chased {other.name}! ({self.following} times)")
```

```python
tom = User("Tom")
jerry = User("Jerry")
```

```python
for i in range(10):
    tom.follow(jerry)
```

```python
print("\nFinal Result:")
print(f"Tom following: {tom.following}")
print(f"Jerry followers: {jerry.followers}")
```
